<h1> Football Predictions Data Analysis <h2/>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('https://projects.fivethirtyeight.com/soccer-api/club/spi_global_rankings.csv')

# Convert league column to category data type
df['league'] = df['league'].astype('category')

# Filter data to top 20 teams
top_20 = df[df['rank'] <= 20]

# Create scatter plot
fig, ax = plt.subplots(figsize=(10, 6))
sc = ax.scatter(x=top_20['off'], y=top_20['def'], s=top_20['spi'], c=top_20['league'].cat.codes, cmap='viridis', alpha=0.7)

# Add labels and title
ax.set_xlabel('Offense Rating')
ax.set_ylabel('Defense Rating')
ax.set_title('FiveThirtyEight International Club Soccer Rankings')

# Add legend for color coding
handles, labels = sc.legend_elements()
legend = ax.legend(handles, top_20['league'], loc='upper right')
legend.set_title('League')

# Show plot
plt.show()
```

The code above performs exploratory data analysis on the FiveThirtyEight International Club Soccer Rankings dataset, which includes rankings of soccer clubs based on their offensive and defensive ratings.

First, the data is loaded using the pandas read_csv() function and stored in a DataFrame called df. The league column is then converted to a category data type to optimize memory usage.

Next, the top 20 teams are filtered using the rank column, which is based on the overall soccer performance of the club. The filtered data is stored in a new DataFrame called top_20.

A scatter plot is then created to visualize the relationship between offensive and defensive ratings of the top 20 teams. The scatter() function from the matplotlib library is used to create the plot, with the x-axis representing the off (offensive) ratings and the y-axis representing the def (defensive) ratings. The spi column is used to determine the size of the points in the scatter plot, with larger points indicating higher overall soccer performance.

The league column is used to color code the points in the scatter plot based on the league that the club belongs to. The cat.codes function is used to convert the categorical data in the league column to numerical codes that can be used for color coding. The cmap parameter is set to viridis to indicate the color palette to be used for the plot, and the alpha parameter is set to 0.7 to control the transparency of the points.

Labels and a title are added to the plot using the set_xlabel(), set_ylabel(), and set_title() functions from the matplotlib library.

Finally, a legend is added to the plot to explain the color coding used for the points. The legend_elements() function is used to create handles and labels for the legend, and the legend() function is used to place the legend in the upper right corner of the plot.

The resulting plot shows a clear clustering of teams based on their offensive and defensive ratings, with larger points indicating higher overall performance. The color coding also allows for easy differentiation of the leagues that each team belongs to. Overall, this analysis provides valuable insights into the current international soccer rankings of top-performing clubs.

Exploratory data analysis (EDA) is the process of analyzing and summarizing data to discover patterns, relationships, and trends. In this case, we will perform EDA on the FiveThirtyEight International Club Soccer Rankings dataset to gain insights into the rankings and the performance of the clubs.

First, let's load the dataset and take a look at its structure and summary statistics:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('https://projects.fivethirtyeight.com/soccer-api/club/spi_global_rankings.csv')

# Display head and summary statistics
print(df.head())
print(df.describe())

```

This code will load the dataset and display the first few rows of data, as well as summary statistics for the numeric columns. We can see that there are 6280 rows and 18 columns in the dataset, with rankings for clubs across different leagues around the world.

Next, let's examine the distribution of the rankings:

```python
# Plot histogram of rankings
plt.hist(df['rank'], bins=20)
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title('Distribution of Club Rankings')
plt.show()
```

This code will create a histogram of the club rankings, with the x-axis representing the rank and the y-axis representing the frequency of clubs at each rank. We can see that the distribution is roughly normal, with a peak around rank 500 and a long tail of lower-ranked clubs.

We can also explore the relationship between different variables, such as offense and defense ratings:

```python
# Plot scatter plot of offense vs defense ratings
plt.scatter(x=df['off'], y=df['def'], alpha=0.3)
plt.xlabel('Offense Rating')
plt.ylabel('Defense Rating')
plt.title('Offense vs Defense Ratings')
plt.show()
```

This code will create a scatter plot of the offense ratings on the x-axis and defense ratings on the y-axis, with the alpha parameter set to 0.3 to add transparency to the points. We can see that there is a generally positive relationship between offense and defense ratings, with higher-rated clubs performing well in both categories.

We can also examine the rankings and ratings of clubs in different leagues:

```python
# Create box plot of offense ratings by league
plt.boxplot([df[df['league'] == league]['off'] for league in df['league'].unique()])
plt.xticks(range(1, len(df['league'].unique()) + 1), df['league'].unique(), rotation=90)
plt.xlabel('League')
plt.ylabel('Offense Rating')
plt.title('Offense Ratings by League')
plt.show()
```

This code will create a box plot of the offense ratings for each league, with the x-axis representing the league and the y-axis representing the offense rating. We can see that there is a wide range of ratings across leagues, with the Spanish Primera Division and the English Premier League having the highest average ratings.

Finally, we can examine the relationship between rank and other variables:

```python
# Plot scatter plot of rank vs offense rating, with color representing league
df['league'] = df['league'].astype('category')

plt.scatter(x=df['rank'], y=df['off'], c=df['league'].cat.codes, cmap='viridis', alpha=0.5)
plt.xlabel('Rank')
plt.ylabel('Offense Rating')
plt.title('Rank vs Offense Rating by League')
plt.show()
```

This code will create a scatter plot of the rank on the x-axis and offense rating on the y-axis, with the color of each point representing the league. We can see that there is a clear separation between the top-ranked clubs and the rest, with most of the top clubs coming from the Spanish Primera Division and the English Premier League.
